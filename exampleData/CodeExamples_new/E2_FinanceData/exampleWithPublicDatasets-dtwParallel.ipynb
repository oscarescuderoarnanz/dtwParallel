```python
# PAQUETE DESARROLLADO: DTW DEPENDIENTE

class Input:
    def __init__(self):
        self.check_errors = False 
        self.type_dtw = "d"
        self.MTS = True
        self.regular_flag = 0
        self.n_threads = -1
        self.local_dissimilarity = "norm2"
        self.visualization = False
        self.output_file = False
        self.DTW_to_kernel = False
        self.sigma = 1
        self.itakura_max_slope = None
        self.sakoe_chiba_radius = None
        
        
x = np.load('../../Data/E2_FinanceData/FinanceData_20days.npy', allow_pickle=True)
x = x[0:252,:,:]
y = x.copy()

print(x.shape)
print(y.shape)

number = 2**7
n_threads = [2**n for n in range(number) if 2**n < number]
n_threads.append(-1)

# Hacer 30 veces y quedarme con la media o mediana del tiempo.
dicc = {}
for i in range(len(n_threads)):
    f = open("contador_dep.txt", "a")
    f.write(str(i))
    f.close()
    times_1 = []
    for j in range(30):
        input_obj = Input()
        input_obj.n_threads = n_threads[i]

        inicio = time.time()

        _ = dtw_functions.dtw_tensor_3d(x, y, input_obj)

        fin = time.time()

        times_1.append(fin-inicio)

    dicc[i] = times_1

    
f = open("./dtwParallel_252.txt", "a")
f.write(str(dicc))
f.close()
```
